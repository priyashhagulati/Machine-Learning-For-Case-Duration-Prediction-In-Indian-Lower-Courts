```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import string
from pathlib import Path

data_dir = Path("/content/drive/MyDrive/ADA_Judiciary")
```

```python
state_district_key = pd.read_csv(data_dir / "keys" / "cases_district_key.csv")
```

```python
state_district_key.head()
```

```python
state_district_key = state_district_key[['state_code', 'state_name', 'dist_code', 'district_name']].drop_duplicates()

# Sort by state_code and district_code in ascending order
state_district_key = state_district_key.sort_values(by=['state_code', 'dist_code'], ascending=True)

# Reset index (optional)
state_district_key = state_district_key.reset_index(drop=True)

# Display the unique rows
print(state_district_key)
```

```python
state_district_key["state_district"] = state_district_key["state_code"].astype(str) + "-" + state_district_key["dist_code"].astype(str)
```

```python
state_district_key.head()
```

```python
state_district_key.to_csv(data_dir / "cleaned" / "state_district_key.csv", index=False)
```
